```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sklearn.metrics as mt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

```python
data = pd.read_csv('data/01_logistic regression.csv')

data.drop(columns=['id', 'Unnamed: 32'], inplace=True)
data['diagnosis'] = (data['diagnosis']=='M').astype(int)
data.head(3)
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
y = data['diagnosis']
X = data.drop(columns='diagnosis')
columns = X.columns     # plot tree | feature importance

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = DecisionTreeClassifier(random_state=0)
model.fit(X_train, y_train)
predictions = model.predict(X_test)

confusion_matrix = mt.confusion_matrix(y_test, predictions)
print('\nConfusion Matrix:\n', confusion_matrix)

accuracy = mt.accuracy_score(y_test, predictions)
print('\nAccuracy:', accuracy)

plt.figure(figsize=(40,20))
plot_tree(model, filled=True, 
          fontsize=14, feature_names=columns)
plt.show()
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_leaf_nodes': range(2,10),
    'max_depth': range(2,10),
    'min_samples_split': range(2,10),
    'min_samples_leaf': range(2,10)
}

grid = GridSearchCV(estimator=model, param_grid=params, cv=10, n_jobs=-1)
grid.fit(X_train, y_train)

print(grid.best_params_)
print(grid.best_score_)
print(grid.best_estimator_)
```

```python
model = grid.best_estimator_
model.fit(X_train, y_train)
predictions = model.predict(X_test)

confusion_matrix = mt.confusion_matrix(y_test, predictions)
print('\nConfusion Matrix:\n', confusion_matrix)

accuracy = mt.accuracy_score(y_test, predictions)
print('\nAccuracy:', accuracy)

plt.figure(figsize=(20,10))
plot_tree(model, filled=True, 
          fontsize=14, feature_names=columns)
plt.show()
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

```python
model = RandomForestClassifier(random_state=0)
model.fit(X_train, y_train)
predictions = model.predict(X_test)

confusion_matrix = mt.confusion_matrix(y_test, predictions)
print('\nConfusion Matrix:\n', confusion_matrix)

accuracy = mt.accuracy_score(y_test, predictions)
print('\nAccuracy:', accuracy)
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [2, 5, 10],
    'min_samples_split': [2, 5, 10],
    'n_estimators': [50, 100, 200, 500, 750, 1000]
}

grid = GridSearchCV(estimator=model, param_grid=params, cv=10, n_jobs=-1)
grid.fit(X_train, y_train)

print(grid.best_params_)
print(grid.best_score_)
print(grid.best_estimator_)
print(grid.best_index_)
```

```python
model = grid.best_estimator_
model.fit(X_train, y_train)
predictions = model.predict(X_test)

confusion_matrix = mt.confusion_matrix(y_test, predictions)
print('\nConfusion Matrix:\n', confusion_matrix)

accuracy = mt.accuracy_score(y_test, predictions)
print('\nAccuracy:', accuracy)


fig, (ax1, ax2) = plt.subplots(2,1, figsize=(30,30))

plot_tree(model[0], filled=True, fontsize=14, ax=ax1)
best_model_index = grid.best_index_
plot_tree(model[best_model_index], filled=True, fontsize=14, ax=ax2)

plt.show()
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
importance = {'feature': model.feature_importances_}
importance = pd.DataFrame(importance, index=columns)
importance.sort_values(by='feature', ascending=True, inplace=True)

plt.figure(figsize=(12,8))
plt.barh(importance.index, importance['feature'])
plt.xticks(np.arange(0.0, 0.14, 0.01))
plt.grid(axis='x', alpha=0.4)
plt.show()
```
