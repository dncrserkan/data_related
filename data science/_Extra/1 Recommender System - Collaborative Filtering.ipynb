```python
import pandas as pd
from sklearn.metrics import pairwise_distances
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

```python
movies = pd.read_csv('data/01_ml-latest-small/movies.csv')
ratings = pd.read_csv('data/01_ml-latest-small/ratings.csv')

display(movies.tail(3), ratings.tail(3))
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
_df = pd.merge(movies, ratings, on='movieId')
_df.head(3)
```

```python
df = _df[['userId', 'title', 'rating']]
df.head(3)
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
pivot_df = df.pivot_table(index='title', columns='userId', values='rating', fill_value=0)
pivot_df.head(3)
print(pivot_df.shape)
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
matrix = pairwise_distances(pivot_df, metric='correlation')
pd.DataFrame(matrix)
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
_index = pivot_df.index.tolist()
index = _index.index('Godfather, The (1972)')
index
```

```python
indexes = (matrix[index].argsort()[1:6])

for i in indexes:
    print(_index[i])
```
