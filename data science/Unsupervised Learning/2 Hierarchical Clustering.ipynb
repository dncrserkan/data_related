```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2 )"> &nbsp; </p>

```python
data = pd.read_csv('data/01_Mall_Customers.csv')
data.head(3)
```

```python
data.drop(columns='CustomerID', inplace=True)
data.isna().sum()
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
X = data[['Age', 'Annual Income (k$)']]

plt.figure(figsize=(4, 3))
sns.scatterplot(X, x='Age', y='Annual Income (k$)', color='black')
plt.show()
```

```python
model  = AgglomerativeClustering()
predictions = model.fit_predict(X)

# print(predictions)
graph = pd.concat([X, pd.Series(predictions, name='Label')], axis=1)

plt.figure(figsize=(4, 3))
sns.scatterplot(graph, x='Age', y='Annual Income (k$)', hue='Label')
plt.legend([],frameon=False)
plt.show()
```

```python
_model  = AgglomerativeClustering(n_clusters=3)
_predictions = _model.fit_predict(X)

_graph = pd.concat([X, pd.Series(_predictions, name='Label')], axis=1)

plt.figure(figsize=(4, 3))
sns.scatterplot(_graph, x='Age', y='Annual Income (k$)', hue='Label')
plt.legend([],frameon=False)
plt.show()
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
dendrogram(linkage(X.sample(20))) # X
plt.xlabel('Data')
plt.ylabel('Distance')
plt.xticks([])
plt.show()
```

```python
# data = pd.read_csv('data/02_Iris.csv', index_col='Id') # column names are different
data = sns.load_dataset('iris')
data.head(3)
```

```python
X = data.drop(columns=['species'])
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
methods = ['single', 'complete', 'average',
           'centroid', 'median', 'ward']
h_clusters = {method:linkage(X, method=method) for method in methods}

fig, axs = plt.subplots(2,3, figsize=(12,8))

for (name, model), (i,j) in zip(h_clusters.items(), np.ndindex(axs.shape)):
    dendrogram(model, ax=axs[i,j])
    axs[i,j].set_title(name, fontweight='bold')
    axs[i,j].set_xticklabels([])
plt.show()
```

```python
model  = AgglomerativeClustering(n_clusters=2, linkage='ward')
predictions = model.fit_predict(X)

graph = pd.concat([X, pd.Series(predictions, name='Label')], axis=1)

plt.figure(figsize=(4, 3))
sns.scatterplot(graph, x='sepal_length', y='sepal_width', hue='Label')
plt.legend([],frameon=False)
plt.show()
```
