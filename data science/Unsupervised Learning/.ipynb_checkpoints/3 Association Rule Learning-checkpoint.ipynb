```python
import pandas as pd

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
from pyECLAT import ECLAT
```

<p style="background-image: linear-gradient(to right, #0D9276, #41b675)"> &nbsp;</p>

```python
data = pd.read_csv('data/02_GroceryStoreDataSet.csv', header=None)
data.columns = ['Products']
data.head(3)
```

<p style="background-image: linear-gradient(#0D9276, #FFFFFF 10%)"> &nbsp; </p>

```python
data = data['Products'].str.split(',')
data.head(3)
```

```python
encoder = TransactionEncoder()
data_en = encoder.fit_transform(data)

data = pd.DataFrame(data_en, columns=encoder.columns_)
data.head()
```

<p style="background-image: linear-gradient(#0D9276, #FFFFFF 10%)"> &nbsp; </p>

```python
df_1 = apriori(data, min_support=0.05, use_colnames=True)
df_2 = association_rules(df_1, metric='confidence', min_threshold=0.5)
df_2
```

```python
data = pd.read_csv('data/02_GroceryStoreDataSet.csv', header=None)
data.columns = ['Products']
data.head(3)
```

```python
temp = data['Products'].str.split(',')
data = pd.DataFrame(temp.to_list())
data.head(3)
```

```python
_1 = min_product = 2
_2 = max_product = max([len(x) for x in temp])
_3 = min_support = 0.2

eclat = ECLAT(data, verbose=True)
# eclat.df_bin
a, b = eclat.fit(min_combination=_1, max_combination=_2, min_support=_3)
display(pd.Series(b))
```

<p style="background-image: linear-gradient(to right, #0D9276, #41b675)"> &nbsp;</p>

```python
data = pd.read_csv('data/03_OnlineRetail.csv')
data.head(3)
```

```python
print(data.isna().sum().sum())
data.dropna(inplace=True)
print(data.isna().sum().sum())
```

```python
print(data.shape)
data = data[~data['InvoiceNo'].str.contains('C')]
print(data.shape)
```

```python
data_UK =data[data['Country']=='United Kingdom'][['InvoiceNo', 'Description', 'Quantity']]
data_UK.head(3)
```

```python
data_UK = data_UK.groupby(['InvoiceNo', 'Description']).sum()
data_UK = data_UK.unstack(fill_value=0)
data_UK.sample(3)
```

```python
# data_UK = data_UK.map(lambda x: 0 if x==0 else 1) # DeprecationWarning
data_UK = data_UK.map(lambda x: False if x==0 else True)
data_UK.sample(3)
```

```python
df_1 = apriori(data_UK, min_support=0.02, use_colnames=True)

df_2 = association_rules(df_1, metric='confidence')
display(df_2)

df_3 = association_rules(df_1, metric='lift', min_threshold=1)
display(df_3)
```
