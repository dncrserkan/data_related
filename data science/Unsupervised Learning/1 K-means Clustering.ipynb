```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

plt.style.use('default')
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

```python
data = pd.read_csv('data/01_Mall_Customers.csv')
data.head(3)
```

```python
data.drop(columns='CustomerID', inplace=True)
data.isna().sum()
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
X = data[['Age', 'Annual Income (k$)']]
```

```python
model = KMeans(n_clusters=2, n_init=10, random_state=0)
fitted = model.fit(X)

clusters = fitted.labels_
centers = model.cluster_centers_


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9,3))

ax1.scatter(X['Age'], X['Annual Income (k$)'])

ax2.scatter(X['Age'], X['Annual Income (k$)'], c=clusters, cmap='winter')
ax2.scatter(centers.T[0], centers.T[1], color='red', s=100, marker='*')

plt.show()
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
wcss = []    # Within-Cluster Sum of Squares
for k in range(1,20):
    _model = KMeans(n_clusters=k, n_init=10, random_state=0)
    _model.fit(X)
    wcss.append(_model.inertia_)

# plt.figure(figsize=(8,3))
plt.plot(range(1, 20), wcss, marker='o')
plt.xlabel('Number of Cluster (K)', fontweight='bold')
plt.ylabel(' Within-Cluster Sum of Squares', fontweight='bold')
plt.xticks(range(1,20))
plt.grid(alpha=0.3)
plt.show()
```

```python
_model = KMeans(random_state=0, n_init=10)
visualizer = KElbowVisualizer(_model, k=(1,20))
visualizer.fit(X)
visualizer.show()
plt.show()
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
k = visualizer.elbow_value_ # 4
model_best = KMeans(n_clusters=k, n_init=10, random_state=0)
fitted_best = model_best.fit(X)

clusters_best = fitted_best.labels_
centers_best = model_best.cluster_centers_

plt.figure(figsize=(4,3))
plt.scatter(X['Age'], X['Annual Income (k$)'], c=clusters_best, cmap='winter_r')
plt.scatter(centers_best.T[0], centers_best.T[1], color='red', s=100, marker='*')
plt.show()
```
