```python
import pandas as pd

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
from pyECLAT import ECLAT
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

```python
data = pd.read_csv('data/04_GroceryStoreDataSet.csv', header=None)
data.columns = ['Products']
data.head(3)
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
data = data['Products'].str.split(',')
data.head(3)
```

```python
encoder = TransactionEncoder()
data_en = encoder.fit_transform(data)

data = pd.DataFrame(data_en, columns=encoder.columns_)
data.head()
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
df_1 = apriori(data, min_support=0.05, use_colnames=True)
df_2 = association_rules(df_1, metric='confidence', min_threshold=0.5)
df_2
```

```python
data = pd.read_csv('data/04_GroceryStoreDataSet.csv', header=None)
data.columns = ['Products']
data.head(3)
```

```python
temp = data['Products'].str.split(',')
data = pd.DataFrame(temp.to_list())
data.head(3)
```

```python
_1 = min_product = 2
_2 = max_product = max([len(x) for x in temp])
_3 = min_support = 0.2

eclat = ECLAT(data, verbose=False) #verbose=True
eclat.df_bin
a, b = eclat.fit(min_combination=_1, max_combination=_2, min_support=_3, verbose=False) #verbose=True
display(pd.Series(b).to_frame())
```
