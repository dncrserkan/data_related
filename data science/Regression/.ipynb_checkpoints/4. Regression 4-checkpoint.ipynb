```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import sklearn.metrics as mt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures, SplineTransformer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection  import train_test_split as tts

import statsmodels.api as sm
import statsmodels.formula.api as smf
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

```python
data = pd.read_csv('data/09_piecewise.csv')
df = data.copy()
data.tail(3)
```

```python
y = df['y']
X = df[['x']]
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.2, random_state=0)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10,3))

sns.scatterplot(x=X_train['x'], y=y_train, alpha=0.3, ax=axs[0])
sns.scatterplot(x=X_test['x'], y=y_test, color='red', alpha=0.3, ax=axs[0])

sns.regplot(x=X_train['x'], y=y_train, ci=None, order=1, 
            line_kws={"color": "black", 'label':'Linear Model'}, 
            scatter_kws={'alpha':0.3}, ax=axs[1])

axs[1].legend(loc='best')
plt.tight_layout()
plt.show()
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

**Polynomial Regression**

```python
pipe = Pipeline([
            ('poly', PolynomialFeatures(degree=2)), 
            ('model', LinearRegression())
])

pipe.fit(X_train, y_train)
predictions = pipe.predict(X_test)
print('R2   :', mt.root_mean_squared_error(y_test, predictions))

model = pipe.named_steps['model']
print(model.intercept_, model.coef_)


sns.lineplot(x=X['x'], y=pipe.predict(X), color='black', label='Polynomial Model')
sns.scatterplot(x=X_train['x'], y=y_train, alpha=0.3, label=None)
sns.scatterplot(x=X_test['x'], y=y_test, color='red', alpha=0.3, label=None)

plt.gcf().set_size_inches(5,3)
plt.legend(loc='best')
plt.show()
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

**Polynomial Spline Regression**

```python
pipe = Pipeline([
            ('polynomial spline', SplineTransformer(n_knots=4, degree=2)), 
            ('model', LinearRegression())
])

pipe.fit(X_train, y_train)
predictions = pipe.predict(X_test)
print('R2   :', mt.root_mean_squared_error(y_test, predictions))

model = pipe.named_steps['model']
print(model.intercept_, model.coef_)


sns.lineplot(x=X['x'], y=pipe.predict(X), color='black', label='Polynomial Spline Model')
sns.scatterplot(x=X_train['x'], y=y_train, alpha=0.3, label=None)
sns.scatterplot(x=X_test['x'], y=y_test, color='red', alpha=0.3, label=None)

splt = SplineTransformer(n_knots=4, degree=2)
splt.fit(X_train)
knots = splt.bsplines_[0].t
plt.vlines(knots[2:-2], ymin=50, ymax=400, linestyles='dashed', linewidth=0.5)

plt.gcf().set_size_inches(5,3)
plt.legend(loc='best')
plt.show()
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

**Linear Spline Regression**

```python
pipe = Pipeline([
            ('polynomial spline', SplineTransformer(n_knots=4, degree=1)), 
            ('model', LinearRegression())
])

pipe.fit(X_train, y_train)
predictions = pipe.predict(X_test)
print('R2   :', mt.root_mean_squared_error(y_test, predictions))

model = pipe.named_steps['model']
print(model.intercept_, model.coef_)

splt = SplineTransformer(n_knots=4, degree=1)
splt.fit(X_train)
knots = splt.bsplines_[0].t
# print(knots)

sns.lineplot(x=X['x'], y=pipe.predict(X), color='black', label='Linear Spline Model')
sns.scatterplot(x=X_train['x'], y=y_train, alpha=0.3, label=None)
sns.scatterplot(x=X_test['x'], y=y_test, color='red', alpha=0.3, label=None)

plt.vlines(knots[1:-1], ymin=50, ymax=400, linestyles='dashed', linewidth=0.5)

plt.gcf().set_size_inches(5,3)
plt.legend(loc='best')
plt.show()
```

**WITH STATSMODELS**

```python
def _plot_it(name=None):
    plt.xlabel('x')
    plt.ylabel('y')
    if name:
        plt.title(name)
    plt.gcf().set_size_inches(5,3)
    plt.tight_layout()
    plt.show()
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Linear Model </p>

```python
formula = f'y ~ x'
models = smf.ols(formula=formula, data=data).fit()
print(models.summary(slim=True))
predictions = models.predict(data)

plt.scatter(data['x'], data['y'], alpha=0.3)
plt.plot(data['x'], predictions, color='black', linewidth=2.5)
_plot_it()
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Polynomial Models </p>

```python
# formula='y ~  X1*X2 + X1 + X2 + X3 + X4'  # Interactions
formula = f'y ~ x + I(x**2)'
models = smf.ols(formula=formula, data=data).fit()
print(models.summary(slim=True))
predictions = models.predict(data)

plt.scatter(data['x'], data['y'], alpha=0.3)
plt.plot(data['x'], predictions, color='black', linewidth=2.5)
_plot_it('Polynomial Model')
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Spline Model </p>

```python
lsm = 'Linear Spline Model'
psm = 'Polynomial Spline Model'
csm = 'Cubic Spline Model'
names = [lsm, psm, csm]
```

```python
formulas, models = {}, {}
for i, name in enumerate(names):
    formulas[name] = f'y ~ bs(x, df=3, degree={i+1})'
    models[name] = smf.ols(formula=formulas[name], data=data).fit()
    print(models[name].summary(slim=True, title=f'{name} - OLS Regression Results'))
    predictions = models[name].predict(data)

    
    plt.scatter(data['x'], data['y'], alpha=0.3)
    plt.plot(data['x'], predictions, color='black', linewidth=2.5)

    # smoothed = sm.nonparametric.lowess(predictions, data['x'], frac=0.5) # frac=0.05 
    # plt.plot(smoothed[:, 0], smoothed[:, 1], color='C1')
    
    _plot_it(name)
```
