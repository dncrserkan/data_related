```python
import pandas as pd
import matplotlib.pyplot as plt

import sklearn.metrics as mt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

```python
data = pd.read_csv('data/06_Position_Salaries.csv')
data.tail(3)
```

```python
y = data['Salary']
X = data[['Level']]
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
dtm = DecisionTreeRegressor(random_state=0)
dtm.fit(X, y)
predictions = dtm.predict(X)

_dtm = DecisionTreeRegressor(max_leaf_nodes=3, random_state=0)
_dtm.fit(X, y)
_predictions = _dtm.predict(X)


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,2))
ax1.scatter(X, y, color='red')
ax1.plot(X, predictions)
ax2.scatter(X, y, color='red')
ax2.plot(X, _predictions)
plt.show()
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
plt.figure(figsize=(15,5), dpi=200)
plot_tree(_dtm, filled=True, fontsize=14,
          feature_names=X.columns, class_names=y)
plt.show()
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
_dtm = DecisionTreeRegressor(max_leaf_nodes=5, random_state=0)
_dtm.fit(X, y)
_predictions = _dtm.predict(X)

plt.figure(figsize=(20,10), dpi=200)
plot_tree(_dtm, filled=True, fontsize=14,
          feature_names=X.columns, class_names=y)
plt.show()
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

```python
data = pd.read_csv('data/07_IceCreamData.csv')
data.tail(3)
```

```python
y = data['Revenue']
X = data[['Temperature']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
dtm = DecisionTreeRegressor(random_state=0)
dtm.fit(X_train, y_train)
predictions = dtm.predict(X_test)

print('R2   :', mt.r2_score(y_test, predictions))
print('RMSE :', mt.root_mean_squared_error(y_test, predictions))
```

```python
_dtm = DecisionTreeRegressor(max_leaf_nodes=20, random_state=0)
_dtm.fit(X_train, y_train)
_predictions = _dtm.predict(X_test)

print('R2   :', mt.r2_score(y_test, _predictions))
print('RMSE :', mt.root_mean_squared_error(y_test, _predictions))
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%)"> &nbsp; </p>

```python
params = {
    'min_samples_split': range(2, 50),
    'max_leaf_nodes': range(2,50)
}

grid = GridSearchCV(estimator=dtm, param_grid=params, cv=10, n_jobs=-1)
grid.fit(X_train, y_train)

print(grid.best_score_)
print(grid.best_params_)
print(grid.best_estimator_)
```

```python
# _dtm = DecisionTreeRegressor(max_leaf_nodes=31, min_samples_split=4, random_state=0)
_dtm = grid.best_estimator_
_dtm.fit(X_train, y_train)
_predictions = _dtm.predict(X_test)

print('R2   :', mt.r2_score(y_test, _predictions))
print('RMSE :', mt.root_mean_squared_error(y_test, _predictions))
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

**Bagging - Decision Tree**

```python
bm = BaggingRegressor(random_state=0)
bm.fit(X_train, y_train)
predictions = bm.predict(X_test)

print('R2   :', mt.r2_score(y_test, predictions))
print('RMSE :', mt.root_mean_squared_error(y_test, predictions))
```

```python
params = {
    'n_estimators': range(2, 25),
}

grid = GridSearchCV(estimator=bm, param_grid=params, cv=10, n_jobs=-1)
grid.fit(X_train, y_train)

print(grid.best_params_)
print(grid.best_score_)
print(grid.best_estimator_)
```

```python
_bm = grid.best_estimator_
_bm.fit(X_train, y_train)
_predictions = _bm.predict(X_test)

print('R2   :', mt.r2_score(y_test, _predictions))
print('RMSE :', mt.root_mean_squared_error(y_test, _predictions))
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

**Random Forest - Decision Tree**

```python
rfm = RandomForestRegressor(random_state=0)
rfm.fit(X_train, y_train)
predictions = rfm.predict(X_test)

print('R2   :', mt.r2_score(y_test, predictions))
print('RMSE :', mt.root_mean_squared_error(y_test, predictions))
```

```python
%%time

params = {
    'max_depth': range(2,20),
    'max_features': range(2,20),
    'n_estimators': range(2,20)
}

grid = GridSearchCV(estimator=rfm, param_grid=params, cv=10, n_jobs=-1)
grid.fit(X_train, y_train)

print(grid.best_params_)
print(grid.best_score_)
print(grid.best_estimator_, '\n')
```

```python
_rfm = grid.best_estimator_
_rfm.fit(X_train, y_train)
_predictions = _rfm.predict(X_test)

print('R2   :', mt.r2_score(y_test, _predictions))
print('RMSE :', mt.root_mean_squared_error(y_test, _predictions))
```
