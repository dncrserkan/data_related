```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
```

<p style="background-image: linear-gradient(to right, #0D9276, #41b675)"> &nbsp;</p>

```python
data = pd.read_csv('data/07_Position_Salaries.csv')
data.tail(3)
```

```python
y = data[['Salary']]        # little different from others
X = data[['Level']]

scaler = StandardScaler()
ss_y = scaler.fit_transform(y)
ss_X = scaler.fit_transform(X)

ss_y = ss_y.ravel()         # now like others
```

```python
svm = SVR()
svm.fit(ss_X, ss_y)
predictions = svm.predict(ss_X)

plt.scatter(ss_X, ss_y, color='C3')
plt.plot(ss_X, predictions, marker='x')
plt.gcf().set_size_inches(5,3)
plt.show()
```

<p style="background-image: linear-gradient(#0D9276, #FFFFFF 10%)"> &nbsp; </p>

```python
svm_l = SVR(kernel='linear')
svm_l.fit(ss_X, ss_y)
predictions_l = svm_l.predict(ss_X)

svm_p = SVR(kernel='poly')
svm_p.fit(ss_X, ss_y)
predictions_p = svm_p.predict(ss_X)

svm_p2 = SVR(kernel='poly', degree=2)
svm_p2.fit(ss_X, ss_y)
predictions_p2 = svm_p2.predict(ss_X)

svm_p5 = SVR(kernel='poly', degree=5)
svm_p5.fit(ss_X, ss_y)
predictions_p5 = svm_p5.predict(ss_X)


plt.figure(figsize=(6,4))
plt.scatter(ss_X, ss_y, color='black')
plt.plot(ss_X, predictions_l, color='C1', label='linear')
plt.plot(ss_X, predictions_p, color='C2', label='poly')
plt.plot(ss_X, predictions_p2, color='C3', label='poly - deg=2')
plt.plot(ss_X, predictions_p5, color='C4', label='poly - deg=5')

plt.legend(loc='best')
plt.show()
```
