**EXPLORATORY DATA ANALYSIS**

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/simple_data.csv')
data.tail(3)
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

**Estimates of Location**

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Mean </p>

```python
# data.mean()
data['Age'].mean()
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Median </p>

```python
data['Age'].median()
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Trimmed Mean </p>

```python
from scipy.stats import trim_mean
trim_mean(data['Age'], 0.1)
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Weighted Mean </p>

```python
np.average(data['Age'], weights=data['Height'])
```

<p style="background-image: linear-gradient(#0aa98f, #FFFFFF 10%); font-weight:bold;"> 
    &nbsp; Weighted Median </p>

```python
import wquantiles
wquantiles.median(data['Age'], weights=data['Height'])
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Mode </p>

```python
# data.mode().loc[0]
data['Age'].mode()
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

**Estimates of Variability**

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Standard Deviation </p>

```python
data['Age'].std()
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Variation </p>

```python
data['Age'].var()
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Quartiles </p>

```python
data.quantile([0.25,0.50,0.75], numeric_only=True)
# data['Age'].quantile([0.25,0.50,0.75])
# data['Age'].quantile(0.50)
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Interquartile Range  </p>

```python
Q1, Q3 = data['Age'].quantile([0.25,0.75])
IQR = Q3 - Q1
IQR
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Median Absolute Deviation - MAD  </p>

```python
from statsmodels import robust
robust.scale.mad(data['Age'])
# abs(data['Age'] - data['Age'].median()).median() / 0.6744897501960817
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Kurtosis  </p>

```python
data.kurtosis(numeric_only=True)
data['Age'].kurtosis()
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Skewness  </p>

```python
data.skew(numeric_only=True)
data['Age'].skew()
```

    &nbsp; GENERAL BRIEF  </p>

```python
data['Gender'].describe()
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Frequency Table & HISTOGRAM  </p>

```python
binned_age = pd.cut(data['Height'], 10)
# binned_age
# binned_age.value_counts()
pd.DataFrame(binned_age.value_counts()).sort_index().T
```

```python
binned_age = pd.qcut(data['Height'], 10)
# binned_age
# binned_age.value_counts()
pd.DataFrame(binned_age.value_counts()).sort_index().T
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(13,3))

ax1.hist(data['Height'], bins=10, edgecolor='white') # histtype='step', density=True
ax1.set_title('Height [cm]', fontweight='bold')

more = np.random.normal(loc=170, scale=10, size=1000)
ax2.hist(more, bins=10, edgecolor='white')
ax2.set_title('Height with more data', fontweight='bold')

sns.histplot(more, bins=20, kde=True, ax=ax3)
# sns.kdeplot(data) # data['Height']
ax3.set_title('Height with same data more bin', fontweight='bold')

plt.show()
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; BOX PLOT  </p>

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,4))
sns.boxplot(data['Height'], color='C1', ax=ax1)
sns.boxplot(data, ax=ax2)

plt.tight_layout()
plt.show()
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; BAR PLOT  </p>

```python
counts = data['Gender'].value_counts() # normalize=True

fig = plt.figure(figsize=(4,3))

plt.bar(counts.index, height=counts, zorder=4)
plt.title('Genders', fontweight='bold')
plt.grid(axis='y', zorder=0)
plt.show()
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; SCATTER PLOT  </p>

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,3))

ax1.scatter(x=data['Height'], y=data['Weight'], s=data['Age'])

scatter = sns.scatterplot(data, x='Height', y='Weight', size='Age', hue='Gender', ax=ax2)
handles, labels = scatter.get_legend_handles_labels()
ax2.legend(handles[1:3], labels[1:3], title='Gender')

ax1.set_title('Height-Weight-Age', fontweight='bold')
ax2.set_title('Height-Weight-Age-Gender', fontweight='bold')
plt.grid(alpha=0.3, zorder=0)
plt.show()
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; CORRELATION & HEATMAP </p>

```python
data.iloc[:,:-1].corr()
```

```python
fig = plt.figure(figsize=(4,3))
sns.heatmap(data.iloc[:,:-1].corr(), 
            annot=True, fmt='.2f', cmap='Blues')
plt.show()
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; HEXEGONAL BINING & CONTOURS </p>

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9,4))


more_height = np.random.normal(loc=170, scale=10, size=1000)
more_weight = np.random.normal(loc=70, scale=10, size=1000)
ax1.hexbin(more_height, more_weight, gridsize=30, cmap='Greens')
ax1.set_xlabel('Height')
ax1.set_ylabel('Weight')

sns.kdeplot(x=data['Height'], y=data['Weight'], ax=ax2)
ax2.set_xlabel('Height')
ax2.set_ylabel('Weight')

plt.show()
```
