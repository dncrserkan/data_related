**DATA ANALYSIS INTRO**

Data analysis is the process of inspecting, cleansing, transforming, and modeling data with the goal of discovering useful information, informing conclusions, and supporting decision-making. The key steps in data analysis include:

> **Data Collection:** Gathering relevant data from various sources, such as databases, surveys, experiments, or sensors.
>
> **Data Cleaning:** Handling missing or inconsistent data by correcting, removing, or imputing values to ensure accuracy.
>
> **Data Exploration:** Using descriptive statistics (like mean, median, or standard deviation) and visualization tools (graphs, charts) to summarize and understand the data's basic characteristics.
>
> **Data Transformation:** Reshaping or converting data into a suitable format for analysis, often by scaling, normalizing, or encoding variables.
>
> **Data Modeling:** Applying statistical or machine learning models to the data to make predictions or uncover relationships.
>
> **Interpretation:** Drawing conclusions and insights based on the results of the analysis, which can inform decisions or suggest further investigation.


<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Descriptive Statistic </p>
    
The goal of descriptive statistics is to gain an overview of the distribution of data sets. Descriptive statistics helps to describe and illustrate data sets. The term descriptive statistics covers statistical methods for describing data using statistical characteristics, charts, graphics or tables.

| | |
|-|-|
| **Location parameters** | mean - median - mode |
| **Dispersion parameters** | range - standard deviation - variance |
| **Tables and Charts** | histogram - bar chart - scatter plot - box plot | 

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Inferantial Statistic </p>
    
Inferential statistics uses various analytical tools to draw conclusions about the population from sample data. For a given hypothesis about the population, inferential statistics uses a sample and gives an indication of the validity of the hypothesis based on the sample collected.

Depending on which statement is to be made about the population or which question is to be answered about the population, different statistical methods or hypothesis tests are used.

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Population and Sample </p>

**Population** is the entire set of individuals, items, or data points that share a common characteristic and are of interest in a statistical study. It is often too large to study in its entirety, so researchers make conclusions about the population by studying a subset of it.

A **sample** is a subset of the population that is selected for study. It is used to make inferences about the population without examining every individual in it.

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Data and Variable </p>

**Data** refers to information that are collected, measured, and recorded.

A **variable** is a characteristic or attribute that can take on different values. It represents a property of the data you’re analyzing and is often the subject of study.
> **Dependent Variables** depend, by some law or rule, on the values of other variables <br>
> **Independent Variables** are not depending on any other variable

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Level of Measurement (Scales of Measurement) </p>

The measurement scale is important because it determines the permissible arithmetic operations and thus specifies the possible statistical tests. The level of measurement of a variable can be either **nominal**, **ordinal** or **metric**. In a nutshell: For nominal variables the values can be differentiated, for ordinal variables the values can be sorted and for metric scale level the distances between the values can be calculated. Nominal and ordinal variables are also called **categorical** variables.

> **Categorical data** <br>
> Data that are represents characteristics. Categorical data can also take on numerical values but those numbers don’t have mathematical meaning.
> > **Nominal** values represent discrete units and are used to label variables that have no quantitative value. <br>
> > **Ordinal** values represent discrete and ordered units. It is therefore nearly the same as nominal data, except that its ordering matters.
> 
> **Numeric (metric) data** <br>
> Data that are expressed on a numeric scale.
> > **Continuous** Data that can take on any value in an interval. <br>
> > **Discrete** Data can only take on certain values that are distinct and separate.  

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; </p>
    
**Mean (Average) [ $\mu - x̄$ ]** 
> Mean is the sum of all data points divided by the total number of points. It represents the central value of a dataset.

**Median**
> The median is the middle value in an ordered dataset. If the number of data points is even, the median is the average of the two middle values.

**Mode** 
> The mode is the values that appear most frequently in a dataset.

**Range**
> The range is the difference between the largest and smallest values in a dataset. It gives a measure of how spread out the data is.

**Interquartile Range**
> The interquartile range is the range between the first quartile (25th percentile) and the third quartile (75th percentile) of a dataset. It measures the spread of the middle 50% of the data, which helps reduce the effect of outliers.

**Standard Deviation [ $\sigma - s$ ]**
> Standard deviation measures how much individual data points deviate from the mean. It gives a sense of the spread or dispersion of the data.

**Variance [ $\sigma^2 - s^2$ ]**
> It is a measure of how far a set of numbers is spread out from their average value and it is the square of the standard deviation.

**Median Absolute Deviation - MAD**
> It is a robust measure of variability that shows how far each data point deviates from the median of the dataset. It is less sensitive to outliers than other measures like the standard deviation because it focuses on the median rather than the mean.

**Kurtosis**
> Kurtosis measures the "tailedness" or the sharpness of the peak of a probability distribution. It indicates how much of the variance in a dataset comes from extreme values (outliers) compared to a normal distribution.
>
>> **Mesokurtic (Kurtosis ≈ 0):** The distribution has a normal level of peakedness, similar to a normal distribution. <br>
>> **Leptokurtic (Kurtosis > 0):** The distribution has a higher peak and fatter tails, indicating more extreme values (outliers). <br>
>> **Platykurtic (Kurtosis < 0):** The distribution has a flatter peak and thinner tails, meaning fewer extreme values.


**Skewness**
> Skewness measures the asymmetry of a distribution around its mean. It shows whether the data are more spread out on one side of the mean than the other.
>
>> **Positive Skew (Right-skewed):** The tail on the right side of the distribution is longer or fatter. The mean is greater than the median. There are a few extremely high values pulling the mean to the right. <br>
>> **Negative Skew (Left-skewed):** The tail on the left side of the distribution is longer or fatter. The mean is less than the median. There are a few extremely low values pulling the mean to the left.

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; CONTINGENCY TABLE </p>

```python
df = data[['clarity', 'cut', 'carat']]
crosstab_counts = df.pivot_table(index='clarity', columns='cut', 
                            aggfunc='count', margins=True, observed=True)
crosstab_counts
```

```python
cp = crosstab_counts.copy()
cp.columns = cp.columns.get_level_values('cut')

cp = cp.astype('float')
cp.loc[:,'Ideal':'Fair'] = cp.loc[:,'Ideal':'Fair'].div(cp['All'], axis=0)
cp['All'] = cp['All'] / cp.iloc[-1,-1] 

cp  # crosstab percentage
```
