**Hypothesis Testing**
> This is a method used to determine the accuracy of a hypothesis within a statistical confidence interval. Typically, a claim about a specific characteristic of a population is tested using a sample.
>
> **1. Formulation of Hypotheses**
> > The first step is the formulation of hypotheses to be tested. This includes the Null hypothesis ($H_0$) and the alternative hypothesis ($H_1$). The null hypothesis generally represents an assumption or the current situation, while the alternative hypothesis usually suggests rejecting the null hypothesis or claiming there is a change.
>
> **2. Selection of Test Statistic**
> > A suitable test statistic is chosen for the test. This statistic is a measurable value used in the collection and analysis of data. The choice of test statistic depends on the nature of the hypothesis, the type of data, and the purpose of the hypothesis test.
>
> **3. Determination of Significance Level**
> > The significance level ($\alpha$ level) of the hypothesis test is determined. The significance level is the probability of accepting that the null hypothesis is wrong. Commonly used significance levels are 0.05 or 0.01.
>
> **4. Calculation of Test Statistic and Determination of P-value**
> > Using the dataset, the test statistic is calculated, and the p-value is determined. The p-value is a measure of the accuracy of the null hypothesis. A low p-value leads to the rejection of the null hypothesis (acceptance of the alternative hypothesis).
>
> **5. Decision Making**
> > Finally, the p-value is compared to the significance level. If the p-value is smaller than the significance level (usually less than 0.05), the null hypothesis is rejected, and the alternative hypothesis is accepted. Otherwise, the null hypothesis is accepted.

**Null Hypothesis**
> The null hypothesis is the basic assumption in a statistical test, often denoted by $H_0$. It represents the claim being tested, usually expressing an assumption or the current situation. It states that the current situation is correct or that there is no effect.
>
> The null hypothesis is used to decide whether it is accepted or rejected during a statistical test. If the data does not sufficiently disprove the null hypothesis, it is accepted. If the data is strong enough to disprove it, the null hypothesis is rejected, and the alternative hypothesis is accepted.
>
> Null Hypothesis  Hypothesis of No Difference  Zero Hypothesis  $H_0$

**Alternative Hypothesis**
> The hypothesis opposing the null hypothesis, generally suggesting the existence of a particular effect, relationship, or change. It is usually denoted by $H_1$. It is accepted if the null hypothesis is rejected.
>
> Alternative Hypothesis  Opposing Hypothesis  Research Hypothesis  $H_1$  $H_a$

**Significance Level**
> The significance level, represented by $\alpha$, is a measure of the acceptable risk of error in hypothesis testing. It is used to determine the critical p-value that will lead to the rejection of the null hypothesis. It can be thought of as the threshold for rejecting the null hypothesis. Typically, it is set at 0.05 (5%) or 0.01 (1%).
> | | |
> |-|-|
> | $\alpha$ | Significance Level |
> | $1 -\alpha$ | Confidence Level |

**One-Tailed and Two-Tailed Tests**
> In a one-tailed hypothesis test, the alternative hypothesis is expressed in only one direction. This suggests that the tested variable has an effect in only one direction.
>
> In a two-tailed hypothesis test, the alternative hypothesis is two-directional, suggesting that the variable could have an effect in either direction.

**P-Value**
> The p-value represents the probability of obtaining the observed data or more extreme results under the assumption that $H_0$ is true. It is determined by the test statistic and compared with the significance level ($$).
> | | |
> |-|-|
> | $p < \alpha$ | $H_0$ is Rejected. |
> | $p  \alpha$ | $H_0$ cannot be Rejected. |

**Type I and Type II Errors**
> If we reject $H_0$ when it is actually true, we make a Type I error. The probability of making this error is $$. 
>
> If $H_0$ is false and we accept it, we make a Type II error. $$ is related to statistical power, representing the ability of a test to detect an actual effect. Thus, Type II error is associated with a lack of statistical power.
>
> Type I and Type II errors are important for evaluating the acceptable risk of error and the reliability of a hypothesis test. Ideally, a balance between these two types of errors should be achieved.

**Test Statistic**
> It is a measure used to evaluate the null hypothesis based on the information obtained from the dataset. The test to be applied is chosen according to the dataset and the hypothesis.
> 
> **Parametric Tests**
> > They assume a specific data distribution and analyze accordingly. Typically, they are used to work with data that fits some assumptions.
> 
> **Nonparametric Tests**
> > They do not make any assumptions about the data distribution or require fewer assumptions. Therefore, they can be used with data that do not meet the assumptions of parametric tests.
> > 
> > Nonparametric tests are especially suitable for situations where the data is ordinal or categorical.
>
> The choice of which test to use depends on the characteristics of your dataset, its distribution, and the nature of your hypotheses. If the assumptions are met, a parametric test should be used. In cases where the assumptions are not met, there is a non-parametric test corresponding to each parametric test.
>
> After selecting the appropriate test and significance value ($\alpha$), the hypothesis test is applied. The $p$-value comes from the applied test statistic.

**Normality Assumption Test**
> These are statistical tests used to determine whether a dataset fits to a normal distribution. The results of parametric tests may not be reliable for datasets that do not fit a normal distribution, so it is important to determine whether the data conforms to normality.
> - We can test whether the distribution resembles a normal distribution based on the skewness and kurtosis values. Different values may be acceptable, but when the values are between $1.5$ (_Tabachnik and Fidell, 2013_), the distribution can be considered normal.
> - To get an idea, we can plot the data on a histogram or Q-Q (quantilequantile) plot.
> - Normality tests can be applied.
> > **Shapiro-Wilk Test** <br>
> > **Kolmogorov-Smirnov Test** <br>
> > **Anderson-Darling Test** <br>
> > **...**
>
> The Shapiro-Wilk test is particularly effective for small and medium-sized sample sizes, while the Kolmogorov-Smirnov test is suitable for large sample sizes.
> The Kolmogorov-Smirnov test and the Anderson-Darling test can also be used to test distributions other than the normal distribution.

**Test of Homogeneity of Variances** - Equality of Variances
> This is an assumption used in statistical tests that aim to determine whether the difference between groups is significant. Homogeneity of variances means that the datasets in different groups show similar variability. If the variances between groups are homogeneous, it suggests that the differences between the groups may indeed arise from the groups themselves.
>
> > **Levene's Test** <br>
> > **Bartlett's Test** <br>
> > **...**
>
> When multiple sample groups are drawn from a population, the variance of each sample group and the population variance will tend to be similar and close to each other. 
If the variances of the samples are close to each other and the homogeneity of variances is ensured, we can assume that the samples are drawn from a population with the same distribution.

```python
import pandas as pd
import numpy as np
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)
```

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Functions to Use </p>

```python
def decision(p, alpha=0.05):
    'acceptance or rejection of the null hypothesis'
    if p < alpha: return 'H0 rejected.'
    else: return 'H0 cannot be rejected.'
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; NORMALITY TESTS </p>

```python
data = stats.norm.rvs(loc=38, size=500)
```

<p style="background-image: linear-gradient(#f87674, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Kurtosis and Skewness  </p>

```python
print('Kurtosis:', stats.kurtosis(data))
print('Skewness:', stats.skew(data))
```

<p style="background-image: linear-gradient(#f87674, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Show on Graph  </p>

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,3))

ax1.hist(data, density=True, alpha=0.4)
sns.kdeplot(data, ax=ax1)

stats.probplot(data, dist='norm', plot=ax2)

plt.show()
```

<p style="background-image: linear-gradient(#f87674, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Kolmogorov-Smirnov Test  </p>

| |  |
|---------|--|
| $H_0$ |  Data is normally distributed |
| $H_1$ |  Data is not normally distributed |

```python
# ks = stats.kstest(veri, cdf='norm') # standard normal distribuion
ks = stats.kstest(data, cdf='norm', args=(data.mean(), data.std()))

# print(ks)
print(f'p-value : {ks.pvalue:.5f}')
print('Decision:', decision(ks.pvalue))
```

<p style="background-image: linear-gradient(#f87674, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Shapiro-Wilk Test  </p>

| |  |
|---------|--|
| $H_0$ |  Data is normally distributed |
| $H_1$ |  Data is not normally distributed |

```python
sw = stats.shapiro(data)

# print(sw)
print(f'p-value : {sw.pvalue:.5f}')
print('Decision:', decision(sw.pvalue))
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

<p style="background-image: linear-gradient(#0aa98f, #ffffff 10%); font-weight:bold;"> 
    &nbsp; HOMOGENEITY of VARIANCES TESTS </p>

```python
data1 = stats.norm.rvs(loc=0, scale=1, size=1000)
data2 = stats.norm.rvs(loc=0, scale=5, size=1000)
data3 = stats.norm.rvs(loc=0, scale=10, size=1000)
```

```python
population = stats.norm.rvs(loc=0, scale=1, size=1000)
sample1 = np.random.choice(population, size=750)
sample2 = np.random.choice(population, size=250)
```

```python
data = pd.read_excel('data/01_variance_test.xlsx')
male = data[data['Gender']=='Male']
female = data[data['Gender']=='Female']
```

<p style="background-image: linear-gradient(#f87674, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Show on Graph  </p>

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 3))

sns.kdeplot(data1, label='data1', ax=ax1)
sns.kdeplot(data2, label='data2', ax=ax1)
sns.kdeplot(data3, label='data3', ax=ax1)

sns.kdeplot(population, label=f'p {population.var():.4f}', ax=ax2)
sns.kdeplot(sample1, label=f's1 {sample1.var():.4f}', ax=ax2)
sns.kdeplot(sample2, label=f's2 {sample2.var():.4f}', ax=ax2)

sns.kdeplot(data['Spending'], label='data', ax=ax3)
sns.kdeplot(male['Spending'], label='male', ax=ax3)
sns.kdeplot(female['Spending'], label='female', ax=ax3)

ax1.legend(loc='best')
ax2.legend(loc='best')
ax3.legend(loc='best')
plt.show()
```

<p style="background-image: linear-gradient(#f87674, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Levene's Test  </p>

| | | 
|-|-|
| $H_0$ |  Variances of the groups are homogeneous. |
| $H_1$ |  Variances of the groups are not homogeneous. |

**Levene's test uses the median as the default value to better respond to the situation where normality is not met.*

```python
p = stats.levene(male['Spending'], female['Spending'], center='mean').pvalue # default center='median'

print(f'p-value : {p:.4f}')
print('Decision:', decision(p=p))
```

<p style="background-image: linear-gradient(#f87674, #ffffff 10%); font-weight:bold;"> 
    &nbsp; Bartlett's Test  </p>

| | | 
|-|-|
| $H_0$ |  Variances of the groups are homogeneous. |
| $H_1$ |  Variances of the groups are not homogeneous. |

```python
p = stats.bartlett(male['Spending'], female['Spending']).pvalue

print(f'p-value : {p:.4f}')
print('Decision:', decision(p=p))
```
