**SAMPLING AND DISTRIBUTIONS**

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

```python
import pandas as pd
import numpy as np
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

**Two tailed test**

```python
perfect = stats.norm.rvs(0, scale=1, size=10000)
perfect.sort()

left = perfect[perfect<-2]
right = perfect[perfect>2]
middle = perfect[perfect>-2]
middle = middle[middle<2]

plt.plot(perfect, stats.norm.pdf(perfect), color='black', linewidth=1.5)
plt.fill_between(y1=0, y2=stats.norm.pdf(left), x=left, color='salmon', alpha=0.6, hatch='+',)
plt.fill_between(y1=0, y2=stats.norm.pdf(right), x=right, color='salmon', alpha=0.6, hatch='+')
plt.fill_between(y1=0, y2=stats.norm.pdf(middle), x=middle, color='lightblue', alpha=0.6, hatch='//')

plt.plot([0, 0], [0, 0.42], color='black', linestyle='--')
plt.plot([-2, -2], [0, 0.2], color='black', linestyle='--')
plt.plot([2, 2], [0, 0.2], color='black', linestyle='--')

plt.text(3, 0.1, f'α/2=0.025', horizontalalignment='center')
plt.text(-3, 0.1, f'α/2=0.025', horizontalalignment='center')
plt.text(0, 0.45, f'1-α=0.95', horizontalalignment='center')

plt.text(-3, -0.02, f'H0 rejected', horizontalalignment='center')
plt.text(3, -0.02, f'H0 rejected', horizontalalignment='center')
plt.text(-2, -0.02, f'-t', horizontalalignment='center')
plt.text(2, -0.02, f't', horizontalalignment='center')
plt.text(0, -0.02, f'0', horizontalalignment='center')

plt.gca().get_xaxis().set_visible(False)
plt.gca().get_yaxis().set_visible(False)

plt.ylim(0, 0.5)
plt.xlim(-4, 4)
plt.title('Two-Tailed Test')

plt.tight_layout()
plt.show()
# plt.savefig('Two-Tailed Test', dpi=1000)
```

<p style="background-image: linear-gradient(to right, #0aa98f, #68dab2)"> &nbsp; </p>

**One tailed test**

```python
perfect = stats.norm.rvs(0, scale=1, size=10000)
perfect.sort()

right = perfect[perfect>1.5]
rest = perfect[perfect<=1.5]

plt.plot(perfect, stats.norm.pdf(perfect), color='black', linewidth=1.5)
plt.fill_between(y1=0, y2=stats.norm.pdf(right), x=right, color='salmon', alpha=0.6, hatch='+')
plt.fill_between(y1=0, y2=stats.norm.pdf(rest), x=rest, color='lightblue', alpha=0.6, hatch='//')

plt.plot([0, 0], [0, 0.42], color='black', linestyle='--')
plt.plot([1.5, 1.5], [0, 0.2], color='black', linestyle='--')

plt.text(2.5, 0.1, f'α=0.05', horizontalalignment='center')
plt.text(0, 0.45, f'1-α=0.95', horizontalalignment='center')

plt.text(2.5, -0.025, f'H0 rejected', horizontalalignment='center')
plt.text(1.5, -0.025, f't', horizontalalignment='center')
plt.text(0, -0.025, f'0', horizontalalignment='center')

plt.gca().get_xaxis().set_visible(False)
plt.gca().get_yaxis().set_visible(False)

plt.ylim(0, 0.5)
plt.xlim(-4, 4)
plt.title('One-Tailed Test')

plt.tight_layout()
plt.show()
# plt.savefig('One-Tailed Test', dpi=1000)
```
